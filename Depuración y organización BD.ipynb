# Cargar y unir bases de datos

```python
import pandas as pd
import numpy as np
import openpyxl as oxl
```

```python
# Base departamentos
depa = pd.read_csv('departamentos - copia.csv', sep=";")
depa.head()
```

```python
# Base economicos 
econo = pd.read_csv('economicos - copia.csv', sep=",")
econo.rename({"Unnamed: 0":"a"}, axis="columns", inplace=True) #Se elimino esta columna
econo.drop(["a"], axis=1, inplace=True)
econo.head()
```

```python
empre = pd.read_csv('empresas - copia.csv', sep = ",", encoding='latin-1')
empre.rename({"Unnamed: 0":"a"}, axis="columns", inplace=True) 
empre.drop(["a"], axis=1, inplace=True)
empre.head()
```

```python
union2 = pd.merge(union1, econo, on='NIT')
union2.head()
```

# Limpieza base de datos

```python
## Identificar los missing values
mv = ['na','--','n/a', 'NoSe', '?','NaN',]
Base = pd.read_csv('Base.CSV', na_values= mv)
Base.rename({"Unnamed: 0":"a"}, axis="columns", inplace=True)
Base.drop(["a"], axis=1, inplace=True)
Base.head()
```

```python
# Número de valores pérdidos
Base.isnull().sum()
```

```python
Base.dtypes
```

<p> Existen distintas formas de reemplazar los valores pérdidos, podemos reemplazarlos con el valor de la media, mediana o moda, para datos númericos, moda para datos string o categóricos, o eliminar el registro </p>

```python
null1 = Base['DIRECCION'].mode()
null2 = Base['CIIU_4'].median()
null3 = Base['Utilidad.operacional'].median()
null4 = Base['Year'].mean()
null5 = Base['Empleados'].median()
null6 = int(Base['Tiene.pagina.web'].mode())
null7 = Base['UVT'].median()

print('\n',null1,null2,null3,null4,null5,null6,null7,'\n')
```

```python
Base['DIRECCION'].fillna(null1, inplace = True)
Base['CIIU_4'].fillna(null2, inplace = True)
Base['Utilidad.operacional'].fillna(null3, inplace = True)
Base['Year'].fillna(null4, inplace = True)
Base['Empleados'].fillna(null5, inplace = True)
Base['Tiene.pagina.web'].fillna(null6, inplace = True)
Base['UVT'].fillna(null7, inplace = True)
Base.isnull().sum()
```

```python
#Eliminar la clumna dirección
Base.drop(columns =["DIRECCION"], inplace = True) 
Base.head()
```

```python
Base.isnull().sum()
```

```python
# Eliminemos índices con valores pérdidos y repetidos
Base.dropna(inplace=True)
Base.drop_duplicates(inplace=True)
Base.describe()
```

# Subset

```python
Meta = Base[Base.MUNI_ID_DPTO == 50 ]
Meta
```

#### Crear una nueva columna que clasifique las empresas

```python
Meta.loc[Meta.UVT <= 23563, 'Clasificacion'] = "Micro" 
Meta.loc[(Meta.UVT > 23563) & (Meta.UVT <= 204995), 'Clasificacion'] = 'Pequeña'
Meta.loc[(Meta.UVT > 204995) & (Meta.UVT <= 1736565), 'Clasificacion'] = 'Mediana'
Meta.loc[Meta.UVT > 1736565, 'Clasificacion'] = 'Grande'
Meta
```
